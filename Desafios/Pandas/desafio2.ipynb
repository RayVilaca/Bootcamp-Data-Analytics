```python
import pydataset as ds
import matplotlib as plt
```

```python
women_ds = ds.data('women')
display(women_ds)
```

1. Construa uma nova representação onde os nomes das colunas estão
em português, e os valores de altura são convertidos para
centímetros e os valores de peso convertidos para quilograma 

```python
# cria novas colunas para realizar a conversão de polegada para cm e libras para kg
women_ds['altura'] = women_ds['height'] * 2.54
women_ds['peso'] = women_ds['weight'] * 0.45

# excluir as colunas antigas
women_ds = women_ds.drop('height', axis=1)
women_ds = women_ds.drop('weight', axis=1)

display(women_ds)
```

```python
# renomeia o nome das colunas
women_ds = women_ds.rename(columns={'height': 'altura', 'weight': 'peso'})

# realiza a conversão de polegada para cm e libras para kg
women_ds['altura'] *= 2.54
women_ds['peso'] *= 0.45

display(women_ds)
```

2. Plote o scatterplot de altura x peso da nova representação

```python
women_ds.plot.scatter('altura', 'peso')
```
