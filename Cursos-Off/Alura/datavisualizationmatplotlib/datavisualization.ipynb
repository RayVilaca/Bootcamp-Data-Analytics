Fazendo as primeiras visualizações

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("monitoramento_tempo.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
df.info()
```

```python
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['data'], df['temperatura'])
plt.title("Temperatura no tempo")
```

```python
fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
```

```python
fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title("Temperatura no momento")
```

```python
fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title("Temperatura no momento")
eixo.set_ylabel("Temperatura")
eixo.set_xlabel("Data")
```

```python
fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title("Temperatura no momento", fontsize=20)
eixo.set_ylabel("Temperatura", fontsize=16)
eixo.set_xlabel("Data", fontsize=16)
```

```python
fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title("Temperatura no momento", fontsize=20)
eixo.set_ylabel("Temperatura", fontsize=16)
eixo.set_xlabel("Data", fontsize=16)
eixo.legend(['temperatura'], loc='lower right', fontsize=14)
```

```python
fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color='slateblue')

eixo.set_title("Temperatura no momento", fontsize=20)
eixo.set_ylabel("Temperatura", fontsize=16)
eixo.set_xlabel("Data", fontsize=16)
eixo.legend(['temperatura'], loc='lower right', fontsize=14)
```
