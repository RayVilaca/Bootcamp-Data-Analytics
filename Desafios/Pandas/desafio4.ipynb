```python
import pydataset as ds
```

```python
bundesliga_ds = ds.data('Bundesliga')
display(bundesliga_ds)
```

```python

bundesliga_ds['TotalGols'] = bundesliga_ds['HomeGoals'] + bundesliga_ds['AwayGoals']

maior_quantidade_gols = bundesliga_ds['TotalGols'].max()
jogos_maior_quantidade_gols = bundesliga_ds.loc[bundesliga_ds['TotalGols'] == maior_quantidade_gols]
display(jogos_maior_quantidade_gols)
```

```python
jogos_maior_quantidade_gols_colunas = jogos_maior_quantidade_gols[['HomeTeam', 'AwayTeam', 'HomeGoals', 'AwayGoals', 'Date']]
display(jogos_maior_quantidade_gols_colunas)
```

```python
sorted_bundesliga_df = bundesliga_ds.sort_values(by='TotalGols', ascending=False)
display(sorted_bundesliga_df)
idxs = sorted_bundesliga_df['TotalGols'] == sorted_bundesliga_df.iloc[0]['TotalGols']
sorted_bundesliga_df = sorted_bundesliga_df[idxs]
display(sorted_bundesliga_df)
```

```python
sorted_bundesliga_df_colunas = sorted_bundesliga_df[['HomeTeam', 'AwayTeam', 'HomeGoals', 'AwayGoals', 'Date']]
display(sorted_bundesliga_df_colunas)
```
