```python
"""
    Essa função retorna o número de linhas e colunas de um dataframe

    Parâmetros:
    df (DataFrame): O dataframe que você deseja saber a quantidade de linhas e colunas.

    Retorna:
    str: texto contendo o número de linhas e colunas.
    """

def num_linhas_colunas(df):
    linhas, colunas = df.shape
    return f"O número de linhas é {linhas} e colunas, {colunas}"
```

### **Desafio 2: Conversão**

**Definição do problema:** Dada a tabela com a média de altura e peso de
mulheres estadunidenses, onde os nomes das colunas estão em inglês
height (altura) e weight (peso) e seus respectivos valores estão nas
seguintes unidades: polegada e libras.

**Dicas:**
* 1 polegada = 2,54 centímetros
* 1 libra = 0,45 quilos
* use o método .plot.scatter()

1. Construa uma nova representação onde os nomes das colunas estão
em português, e os valores de altura são convertidos para
centímetros e os valores de peso convertidos para quilograma. 

```python
women_df = pydataset.data('women')

# renomeia o nome das colunas
women_df = women_df.rename(columns={'height': 'altura', 'weight': 'peso'})

# converte polegadas para centímetros e libras para quilogramas
women_df['altura'] *= 2.54
women_df['peso'] *= 0.45

display(women_df)
```

2. Plote o scatterplot de altura x peso da nova representação.

```python
women_df.plot.scatter('altura', 'peso', title='Altura x Peso', xlabel='Altura (cm)', ylabel='Peso (kg)')
```

### **Desafio 3: Titanic**

**Definição do problema:** Dado que você está trabalhando com um grupo
de historiadores que estão analisando o naufrágio do Titanic, informe aos
historiadores as seguintes questões?

**Dicas:**
* value_counts() (Itens 1 e 2)
* seleção de subconjuntos (Itens 3 e 4)

1. Quantas pessoas sobreviveram e quantas morreram?

```python
titanic_df = pydataset.data('titanic')
contagem_coluna_survived = titanic_df['survived'].value_counts()
print(f"No naufrágio do Titanic, {contagem_coluna_survived['yes']} pessoas sobreviveram e {contagem_coluna_survived['no']} pessoas morreram.")
```

2. Quantas pessoas eram da primeira, segunda e terceira classe?

```python
contagem_coluna_class = titanic_df['class'].value_counts()
print(f"Na primeira classe estavam {contagem_coluna_class['1st class']} pessoas")
print(f"Na segunda classe estavam {contagem_coluna_class['2nd class']} pessoas")
print(f"Na terceira classe estavam {contagem_coluna_class['3rd class']} pessoas")
```

3. Qual a porcentagem de sobreviventes na primeira classe?

```python
contagem_class_survived = titanic_df[['class', 'survived']].value_counts()
sobreviventes_1st_class = contagem_class_survived['1st class']['yes'] * 100 / contagem_class_survived['1st class'].sum()
print(f"A porcentagem de sobreviventes na primeira classe é {sobreviventes_1st_class:.2f}%")
```

```python
groupby_class_survived = titanic_df.groupby(['class', 'survived']).size()
sobreviventes_1st_class_groupby = groupby_class_survived['1st class']['yes'] * 100 / groupby_class_survived['1st class'].sum()
print(f"A porcentagem de sobreviventes na primeira classe é {sobreviventes_1st_class_groupby:.2f}%")
```

4. Qual a porcentagem de mortos da terceira classe?

```python
mortos_3rd_class = contagem_class_survived['3rd class']['no'] * 100 / contagem_class_survived['3rd class'].sum()
print(f"A porcentagem de mortos da terceira classe é {mortos_3rd_class:.2f}%")
```

```python
mortos_3rd_class_groupby = groupby_class_survived['3rd class']['no'] * 100 / groupby_class_survived['3rd class'].sum()
print(f"A porcentagem de mortos da terceira classe é {mortos_3rd_class_groupby:.2f}%")
```

### **Desafio 3: Futebol**

**Definição do problema:** Dado que você está trabalhando numa emissora
de televisão, o narrador precisa informar o jogo com o maior número de
gols na história da Bundesliga (Campeonato Alemão). Você precisa fazer
uma busca em todos os jogos da história deste campeonato e verificar qual
jogo é esse e passar a iformação ao narrador, para posteriormente, o
narrador informar aos telespectadores.

**Dicas:**
* Criar uma nova variável "TotalGols" com o total de gols da partida;
* Use o método.max()

1. Verificar os jogos com maior quantidade de gols na história da
Bundesliga

```python
jogos_df = pydataset.data('Bundesliga')

# cria uma coluna com o total de gols marcados no jogo
jogos_df['TotalGoals'] = jogos_df['HomeGoals'] + jogos_df['AwayGoals']

# varifica a maior quantidade de gols marcados nos jogos
maior_quantidade_gols = jogos_df['TotalGoals'].max()

# busca pelos jogos que o total de gols marcados é igual ao número máximo de gols
jogos_maior_quantidade_gols = jogos_df.loc[jogos_df['TotalGoals'] == maior_quantidade_gols]

display(jogos_maior_quantidade_gols)
```

2. Informar a data, os times participantes e o placar do jogo

```python
jogos_maior_quantidade_gols_info = jogos_maior_quantidade_gols[['HomeTeam', 'AwayTeam', 'HomeGoals', 'AwayGoals', 'Date']]
display(jogos_maior_quantidade_gols_info)
```

### **Desafio 3: Forbes**

**Definição do problema:** Você está trabalhando em um jornal e o editor
quer que você responda as seguintes questões:

**Dicas:**
* Método .max()
* Método nlargest()
* Método .groupby()

1. Qual a empresa mais valiosa da lista da Forbes 2000?

```python
forbes_df = pydataset.data('Forbes2000')

maior_market_value = forbes_df['marketvalue'].max()
empresas_mais_valiosas = forbes_df.loc[forbes_df['marketvalue'] == maior_market_value]
print(f"A empresa mais valiosa da lista da Forbes 2000 é {empresas_mais_valiosas.iloc[0]['name']}")

forbes_df_ordenada_marketvalue = forbes_df.sort_values(by='marketvalue', ascending=False)
print(f"A empresa mais valiosa da lista da Forbes 2000 é {forbes_df_ordenada_marketvalue.iloc[0]['name']}")

empresa_mais_valiosa = forbes_df.nlargest(1, 'marketvalue')
print(f"A empresa mais valiosa da lista da Forbes 2000 é {empresa_mais_valiosa.iloc[0]['name']}")
```

2. Construa uma tabela com as TOP 10 empresas mais lucrativas da
lista.

```python
forbes_top10_lucrativa = forbes_df.nlargest(10, 'profits')
display(forbes_top10_lucrativa)
```

```python
forbes_lucrativa_ordenada = forbes_df.sort_values(by='profits', ascending=False)
forbes_top10_lucrativa_ordenada = forbes_lucrativa_ordenada.iloc[0:10]
display(forbes_top10_lucrativa_ordenada)
```

3. Qual a média de valores de empresa das cinco categorias mais
valiosas?

```python
groupby_marketvalue = forbes_df.groupby('category')['marketvalue'].sum()
marketvalue_5catecorias_valiosas = groupby_marketvalue.nlargest(5)
media_marketvalue_5catecorias_valiosas = marketvalue_5catecorias_valiosas.mean()
print(f"A média de valores de empresas das cinco categorias mais valiosas é de {media_marketvalue_5catecorias_valiosas:.2f}")
```
